```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime


# GET to request htm & link url
url = 'https://www.ccny.cuny.edu/registrar/fall'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc)

#Find Table & Extract
table = soup.find("table")

data = []

rows= table.find_all("tr")[1:]
for row in rows:
    cells = row.find_all("td")
    if len(cells) == 3:
        date_text = cells[0].get_text(strip=True) 
        dow_text = cells[1].get_text(strip=True)
        description = cells[2].get_text(strip=True)

        if "January" not in date_text:
            date_text += ", 2021"

        if"-" in date_text:
            month_day_range, year = date_text.rsplit(",", 1)
            month, days = month_day_range.split(" ",1)
            start_day, end_day = days.split("-")
        

            start_date_str= f"{month} {start_day}, {year.strip()}"
            end_date_str= f"{month} {end_day}, {year.strip()}"

        
            start_date = datetime.strptime(start_date_str, "%B %d , %Y").date()
            end_date= datetime.strptime(end_date_str, "%B %d, %Y").date()
            date_text=f"{start_date} - {end_date}"

        else:
            
            date_text = datetime.strptime(date_text,"%B %d, %Y").date()
        
        data.append({"date":date_text, "Dow": dow_text, "Text Description": description})
        
df = pd.DataFrame(data)
df.set_index("date", inplace=True)
pd.set_option('display.max_colwidth', 30)

print(df)

```
